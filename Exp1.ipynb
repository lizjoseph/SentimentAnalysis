```python
disk_engine = create_engine('sqlite:///tweets.db')
```

```python
start = dt.datetime.now()
chunksize = 20000
j = 0
index_start = 1

for df in pd.read_csv('data', chunksize=chunksize, iterator=True, encoding='utf-8',names=[ "ID" , "CONTENT"  , "CREATEDAT" , "RETWEET" , "RTID" , "RTUSERNAME" , "OPEN_DATE" , "USERID" , "USERNAME" , "FULLNAME" , "DESCRIPTION" , "GEO_ENABLED" , "LATITUDE" , "LONGITUDE"  , "LOCATION" , "TIMEZONE" , "LANGUAGE" , "FOLLOWERS" , "FRIENDS" , "RTFOLLOWERS" , "TRANSLATOR" , "STATUS_COUNT" , "PROFILE_IMAGE" , "BACKGROUND_IMAGE" , "BANNER_IMAGE" , "IS_FRIEND" , "IS_FOLLOWER" , "IS_DEFAULT_PROFILE" , "SIZE" , "LAST_UPDATED" , "INREPLYTO" , "SOURCE" ]
):
    df = df.rename(columns={c: c.replace(' ', '') for c in df.columns})
    
    #df['CREATEDAT'] = pd.to_datetime(df['CREATEDAT'])
    #df['OPEN_DATE'] = pd.to_datetime(df['OPEN_DATE'])
    colums=[  "CONTENT"  , "CREATEDAT", "OPEN_DATE" , "USERID" , "USERNAME" , "FULLNAME" , "DESCRIPTION" ,"GEO_ENABLED"  , "LOCATION" , "TIMEZONE" 
            , "FOLLOWERS" ,"STATUS_COUNT" , "PROFILE_IMAGE" , "BACKGROUND_IMAGE" , "BANNER_IMAGE" ,"IS_DEFAULT_PROFILE" , 
            "LAST_UPDATED"  ]
    
    for c in df.colums
        if c not in colums
            df=df.drop(c,axis=1)
    
    j+=1
    print '{} seconds: completed {} rows'.format((dt.datetime.now() - start).seconds, j*chunksize)

    df.to_sql('data', disk_engine, if_exists='append')
    index_start = df.index[-1] + 1
    
```

```python
df = pd.read_sql_query('SELECT TIMEZONE, COUNT(*) as `number_of_zones`'
                       'FROM data '
                       'GROUP BY TIMEZONE '
                       'ORDER BY -number_of_zones', disk_engine)

display(df)
```

```python

df = pd.read_sql_query('SELECT TIMEZONE, COUNT(*) as `zonescount`'
                       'FROM data '
                       'GROUP BY TIMEZONE '
                       'ORDER BY -zonescount', disk_engine)

py.iplot([Bar(x=df.TIMEZONE, y=df.zonescount)], filename='timezonevstatuscount')
```

```python
df = pd.read_sql_query('SELECT USERNAME, STATUS_COUNT, COUNT(STATUS_COUNT) as `statuscount`'
                       'FROM data '
                       'GROUP BY USERNAME '
                       'ORDER BY -statuscount', disk_engine)
display(df)
```

```python

df = pd.read_sql_query('SELECT USERNAME, STATUS_COUNT, COUNT(STATUS_COUNT) as `statuscount`'
                       'FROM data '
                       'GROUP BY USERNAME '
                       'ORDER BY -statuscount', disk_engine)



trace = Scatter(
    x = df.USERNAME,
    y = df.statuscount
)

data = [trace]


py.iplot(data, filename='useridvstatuscount')
```

```python
df = pd.read_sql_query('SELECT USERNAME, FRIENDS, CAST(data.FRIENDS as int64) as numbers '
                       'FROM data ', disk_engine)

display(df)
```

```python
df = pd.read_sql_query('SELECT USERNAME, FRIENDS, IS_FRIEND, MAX(FRIENDS) as `checkFriend`'
                       'FROM data '
                       'GROUP BY USERNAME '
                       'ORDER BY -checkFriend', disk_engine)


xData = df.IS_FRIEND
yData = df.FRIENDS
zData = df.USERNAME
trace = Scatter(
    x = yData,
    y = xData,
    mode = 'markers'
)

layout = Layout(
    xaxis=dict(
        range=[0, df.checkFriend]
    ),
    yaxis=dict(
        range=[0, 1]
    )
    
)

data = [trace]

fig = Figure(data = data, layout= layout)




py.iplot(fig, filename='useridvisfriend')
```
