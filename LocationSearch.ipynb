```python
import numpy as np
import pandas as pd
import datetime as dt
import sqlalchemy
from sqlalchemy import create_engine 


start = dt.datetime.now()

disk_engine = create_engine(r'sqlite:///E:\experiment_tweets_shortest\files1.db')

chunksize = 100

j = 0
index_start = 1

for df in pd.read_csv('data', chunksize=chunksize, iterator=True, encoding='LATIN1', error_bad_lines = False, names= ['ID', "Content","CreateDate","RETWEET", "RTID", "RTUSERNAME", "OPEN_DATE",
                                                      "USERID", "USERNAME", "FULLNAME","DESCRIPTION", "GEO_ENABLED", 
                                                      "LATITUDE", "LONGITUDE", "LOCATION" , "TIMEZONE",  "LANGUAGE", "FOLLOWERS",
                                                      "FRIENDS" , "RTFOLLOWERS", "TRANSLATOR", "STATUS_COUNT", "PROFILE_IMAGE",
                                                      "BACKGROUND_IMAGE", "BANNER_IMAGE", "IS_FRIEND", "IS_FOLLOWER",
                                                      "IS_DEFAULT_PROFILE", "SIZE" , "LAST_UPDATED" , "INREPLYTO" ,"SOURCE"]):
    
    df = df.rename(columns={c: c.replace(' ', '') for c in df.columns}) # Remove spaces from columns 

    df.index += index_start
    j+=1
    print '{} seconds: completed {} rows'.format((dt.datetime.now() - start).seconds, j*chunksize)

    df.to_sql('data', disk_engine, if_exists='append')
    index_start = df.index[-1] + 1

```

```python
import pandas as pd
import sqlalchemy
from sqlalchemy import create_engine 
disk_engine = create_engine(r'sqlite:///E:\experiment_tweets_shortest\files1.db')
df = pd.read_sql_query('Select ID, CONTENT '
                       'FROM data '
                       'LIMIT 100', disk_engine)
df.head()
```

```python
df = pd.read_sql_query('SELECT TIMEZONE, COUNT(*) as `zonescount`'
                       'FROM data '
                       'GROUP BY TIMEZONE '
                       'ORDER BY -zonescount', disk_engine)

py.iplot([Bar(x=df.TIMEZONE, y=df.zonescount)], filename='TweetsTimeZone')
```

```python
df = pd.read_sql_query('SELECT USERNAME, COUNT(*) as `Number_Of_Tweets`'
                       'FROM data '
                       'GROUP BY USERNAME '
                       'ORDER BY -Number_Of_Tweets', disk_engine)

py.iplot([Bar(x=df.USERNAME, y=df.Content)], filename='UsersTweets')
```

```python
#  this query give the number of tweets per location
df = pd.read_sql_query('SELECT Location, COUNT(*) as `NumberOfTweets` '
                        'FROM data '
                        'GROUP BY `Location` '
                       'COLLATE NOCASE '
                       'ORDER BY -NumberOfTweets '
                       'LIMIT 20 ', disk_engine)
df.head()
```

```python
# this code remove the tweets where the location is not define
LOCATIONS = list(df.Location)
LOCATIONS.remove(None)
```

```python
traces = [] # the series in the graph - one trace for each city

for LOCATION in LOCATIONS:
    df = pd.read_sql_query('SELECT TweetsType, COUNT(*) as `NumberOfTweets` '
                           'FROM data '
                           'WHERE LOCATION = "{}" COLLATE NOCASE '
                           'GROUP BY `TweetsType` '
                           'ORDER BY -NumberOfTweets'.format(LOCATION), disk_engine)

    traces.append(Bar(x=df['TweetsType'], y=df.NumberOfTweets, name=LOCATION.capitalize()))
```
